```python
import matplotlib.pyplot as plt
import numpy as np

daily_prices = [115.2, 116.0, 116.6, 119.1, 122.7, 123.1, 122.9, 122.3, 123.8, 124.4, 121.8, 123.2, 122.4, 121.8, 127.9, 127.8, 128.7, 126.7, 128.2, 131.9, 131.0, 132.0, 136.7, 134.9, 133.7, 132.7, 129.4, 131.0, 126.6, 130.9, 132.1, 129.0, 128.8, 130.9, 128.9, 127.1, 127.8, 132.0, 136.9, 139.1, 142.9, 143.2, 142.1, 137.1, 132.0, 134.1, 135.0, 133.9, 137.4, 136.8, 136.9, 136.0, 135.4, 135.1, 135.4, 133.2, 130.8, 129.7, 129.9, 126.0, 125.9, 125.4, 121.0, 121.3, 127.8, 125.1, 122.1, 120.1, 121.4, 116.4, 121.1, 120.0, 122.0, 121.0, 124.0, 125.6, 124.8, 120.5, 120.0, 123.4, 122.5, 120.1, 120.6, 121.2, 121.4, 119.9, 122.2, 123.0, 125.9, 126.2, 127.9, 130.4, 133.0, 131.2, 134.4, 132.0, 134.5, 134.2, 134.8, 133.1, 133.5, 131.9, 134.3, 134.7, 134.4, 133.6, 133.5, 131.5, 132.5, 127.9, 128.1, 129.7, 130.2, 126.9, 125.9, 122.8, 125.0, 127.5, 126.3, 124.9, 124.7, 127.3, 125.4, 127.1, 126.9, 126.9, 125.3, 124.6, 124.3, 125.1, 123.5, 125.9, 125.9, 126.7, 127.1, 126.1, 127.4, 130.5, 129.6, 130.2, 131.8, 130.5, 132.3, 134.0, 133.7, 133.4, 133.1, 134.8, 136.3, 137.0, 137.3, 140.0, 142.0, 144.6, 143.2, 145.1, 144.5, 145.6, 149.2, 148.5, 146.4, 142.5, 146.2, 145.4, 146.8, 148.6, 149.0, 146.8, 145.0, 145.6, 145.9, 145.5, 147.4, 147.0, 147.1, 146.1, 146.1, 145.6, 145.9, 148.9, 149.1, 151.1, 150.2, 146.4, 146.7, 148.2, 149.7, 149.6, 148.4, 147.5, 148.6, 153.1, 151.8, 152.5, 153.7, 154.3, 156.7, 155.1, 154.1, 149.0, 149.6, 148.1, 149.0, 148.8, 146.1, 142.9, 143.4, 145.9, 146.8, 146.9, 145.4, 141.9, 142.8, 141.5, 142.7, 139.1, 141.1, 142.0, 143.3, 142.9, 142.8, 141.5, 140.9, 143.8, 144.8, 146.6, 148.8, 149.3, 149.5, 148.7, 148.6, 149.3, 148.9, 152.6, 149.8, 149.0, 150.0, 151.5, 151.0, 151.3, 150.4, 150.8, 147.9, 147.9, 150.0, 150.0, 151.0, 153.5, 157.9, 160.6, 161.0, 161.4, 161.9, 156.8, 160.2, 165.3, 164.8, 163.8, 161.8, 165.3, 171.2, 175.1, 174.6, 179.5, 175.7, 174.3, 179.3, 172.3, 171.1, 169.8, 173.0, 175.6, 176.3, 180.3, 179.3, 179.4, 178.2, 177.6, 182.0, 179.7, 174.9, 172.0, 172.2, 172.2, 175.1, 175.5, 172.2, 173.1, 169.8, 166.2, 164.5, 162.4, 161.6, 159.8, 159.7, 159.2, 170.3, 174.8, 174.6, 175.8, 172.9, 172.4, 171.7, 174.8, 176.3, 172.1, 168.6, 168.9, 172.8, 172.6, 168.9, 167.3, 164.3, 160.1, 162.7, 164.9, 165.1, 163.2, 166.6, 166.2, 163.2, 159.3, 157.4, 163.0, 158.5, 154.7, 150.6, 155.1, 159.6, 160.6, 164.0, 165.4, 168.8, 170.2, 174.1, 174.7, 175.6, 179.0, 177.8, 174.6, 174.3, 178.4, 175.1, 171.8, 172.1, 170.1, 165.8, 167.7, 170.4, 165.3, 165.1, 167.4, 167.2, 166.4, 161.8, 162.9, 156.8, 156.6, 163.6, 157.7, 158.0, 159.5, 166.0, 156.8, 157.3, 152.1, 154.5, 146.5, 142.6, 147.1, 145.5, 149.2, 140.8, 137.4, 137.6, 143.1, 140.4, 140.5, 143.8, 149.6, 148.8, 148.7, 151.2, 145.4, 146.1, 148.7, 148.0, 142.6, 137.1, 131.9, 132.8, 135.4, 130.1, 131.6, 135.9, 135.4, 138.3, 141.7, 141.7, 137.4, 139.2, 136.7, 138.9, 141.6, 142.9, 146.4, 147.0, 144.9, 145.9, 145.5, 148.5, 150.2, 147.1, 151.0, 153.0, 155.4, 154.1, 153.0, 151.6, 156.8, 157.4, 162.5, 161.5, 160.0, 166.1, 165.8, 165.4, 164.9, 164.9, 169.2, 168.5, 172.1, 173.2, 173.0, 174.6, 174.2, 171.5, 167.6, 167.2, 167.5, 170.0, 163.6, 161.4, 158.9, 157.2, 158.0, 155.8, 154.5, 156.0, 154.5, 157.4, 163.4, 153.8, 155.3, 152.4, 150.7, 154.5, 156.9, 153.7, 152.7, 150.4, 150.8, 151.8, 149.8, 142.5, 138.2, 142.5, 146.1, 146.4, 145.4, 140.1, 140.4, 139.0, 138.3, 143.0, 138.4, 142.4, 143.8, 143.9, 143.4, 147.3, 149.5, 152.3, 149.4, 144.8, 155.7, 153.3, 150.7, 145.0, 138.9, 138.4, 138.9, 139.5, 134.9, 146.9, 149.7, 148.3, 150.0, 148.8, 150.7]

def moving_average(data, n):
    ma = []
    for i in range(len(data)):
        if i < n - 1:
            ma.append(None)
        else:
            ma.append(sum(data[i-n+1:i+1]) / n)
    return ma

def plot_stock_data(data, ma):
    plt.figure(figsize=(12, 6))
    plt.plot(range(1, len(data) + 1), data, label='Daily Price')
    plt.plot(range(1, len(ma) + 1), ma, label=f'{n}-day Moving Average')
    plt.xlabel('Days')
    plt.ylabel('Price')
    plt.title('Stock Price and Moving Average')
    plt.legend()
    plt.grid(True)
    plt.show()

def compare_price_to_ma(data, ma):
    signals = []
    for price, avg in zip(data, ma):
        if avg is not None:
            if price > avg:
                signals.append('+')
            else:
                signals.append('-')
        else:
            signals.append('')
    return signals

# 이동평균선 계산 (예: 20일)
n = 20
ma_values = moving_average(daily_prices, n)

# 그래프 그리기
plot_stock_data(daily_prices, ma_values)

# 매수/매도 신호 분석
signals = compare_price_to_ma(daily_prices, ma_values)

# 결과 출력
plus_count = signals.count('+')
minus_count = signals.count('-')
print(f"매수 신호 (+) 일수: {plus_count}")
print(f"매도 신호 (-) 일수: {minus_count}")
```
